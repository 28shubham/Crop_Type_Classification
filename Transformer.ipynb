```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Input, Dense, Dropout, Attention, LayerNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
import os
import matplotlib.pyplot as plt

# Function to load and preprocess data from a single CSV file
def load_and_preprocess_data(file_path):
    # Load data
    data = pd.read_csv(file_path)

    # Split the dataset into features (spectral bands) and labels (flags)
    X = data.iloc[:, :-1].values  # Features (spectral bands)
    y = data.iloc[:, -1].values  # Labels (flags)

    # Normalize the features (spectral bands) using StandardScaler
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    return X_scaled, y

# Function to process a folder for a specific crop type
def process_crop_type(root_folder, crop_type, is_test=False):
    # Define folder paths
    train_folder_path = os.path.join(root_folder, crop_type, "train")
    validate_folder_path = os.path.join(root_folder, crop_type, "valid")
    test_folder_path = os.path.join(root_folder, crop_type, "test")

    # Process the training dataset
    train_accuracy, val_accuracy, train_loss, val_loss = process_folder(train_folder_path, is_test)

    # Plot training and validation accuracy
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(train_accuracy) + 1), train_accuracy, label='Training Accuracy')
    if val_accuracy:
        plt.plot(range(1, len(val_accuracy) + 1), val_accuracy, label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training and Validation Accuracy - Crop Type {}'.format(crop_type))
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(train_loss) + 1), train_loss, label='Training Loss')
    if val_loss:
        plt.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss - Crop Type {}'.format(crop_type))
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Process the validation dataset
    val_accuracy, _, val_loss, _ = process_folder(validate_folder_path, is_test=False)

    # Process the test dataset
    test_accuracy, _, test_loss, _ = process_folder(test_folder_path, is_test=True)

    # Print final accuracy and loss on the test dataset
    print("Final accuracy on the test dataset for Crop Type {}: {}".format(crop_type, test_accuracy[-1]))
    print("Final loss on the test dataset for Crop Type {}: {}".format(crop_type, test_loss[-1]))

# Function to process a folder
def process_folder(folder_path, is_test=False):
    X_batches = []
    y_batches = []

    for tile_folder in os.listdir(folder_path):
        tile_folder_path = os.path.join(folder_path, tile_folder)
        if os.path.isdir(tile_folder_path):
            file_paths = [os.path.join(tile_folder_path, filename) for filename in os.listdir(tile_folder_path) if filename.endswith(".csv")]
            if file_paths:
                for file_path in file_paths:
                    if os.path.basename(file_path) == "dates.csv":
                        print("Skipping file:", file_path)
                        continue

                    X, y = load_and_preprocess_data(file_path)
                    X_batches.append(X)
                    y_batches.append(y)

    if not X_batches:  # Add this condition to check if there's no data
        print(f"No data found in folder: {folder_path}")
        return [], [], [], []  # Return default values

    X_data = np.concatenate(X_batches)
    y_data = np.concatenate(y_batches)

    # Define the Attention-based Transformer model
    inputs = Input(shape=(X_data.shape[1],))
    dense1 = Dense(128, activation='relu')(inputs)
    dropout1 = Dropout(0.3)(dense1)
    dense2 = Dense(64, activation='relu')(dropout1)
    attention_output = Attention()([dense2, dense2])
    attention_output = LayerNormalization(epsilon=1e-6)(attention_output)
    dense3 = Dense(32, activation='relu')(attention_output)
    dropout2 = Dropout(0.3)(dense3)
    outputs = Dense(4, activation='softmax')(dropout2)

    model = Model(inputs=inputs, outputs=outputs)

    # Compile the model
    optimizer = Adam(learning_rate=0.0001)  # Adjusted learning rate
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

    # One-hot encode the target data
    y_data_encoded = to_categorical(y_data)

    # Train the model
    if not is_test:
        history = model.fit(X_data, y_data_encoded, epochs=5, batch_size=64, verbose=0, validation_split=0.2)
    else:
        history = model.fit(X_data, y_data_encoded, epochs=5, batch_size=64)

    tf.keras.backend.clear_session()
    del model

    # Return training accuracy, validation accuracy, training loss, and validation loss if available
    train_accuracy = history.history['accuracy']
    val_accuracy = history.history.get('val_accuracy', [])
    train_loss = history.history['loss']
    val_loss = history.history.get('val_loss', [])

    return train_accuracy, val_accuracy, train_loss, val_loss

# Define base path for the dataset
root_folder = "/workspace/awadh/nvidia/Vaibhav/Satellite_TSS2/CropWiseSimplified"

# Process each crop type
for i in range(16):
    crop_type = str(i)
    process_crop_type(root_folder, crop_type)
```
